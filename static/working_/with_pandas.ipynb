```python
from time import time
import pandas as pd
import numpy as np
```

```python
df =  pd.read_csv('https://raw.githubusercontent.com/sanyathisside/Predicting-Heart-Disease-using-Machine-Learning/master/heart-disease.csv')
```

```python
from time import time
before = time()
df.describe().loc['mean']
after = time()
print(after - before)
```

```python
from time import time
before = time()
df.mean(axis=0)
after = time()
print(after - before)
```

```python
from time import time
before = time()
df.apply(np.mean)
after = time()
print(after - before)
```

```python
from time import time
before = time()
df.apply('mean')
after = time()
print(after - before)
```
